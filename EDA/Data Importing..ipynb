```python
import pandas as pd

# Specify the file path
file_path = '/home/borna/DC/DCIAProject/Data/jads_data.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Display the DataFrame
df


```

The dataset contains 132,971 entries and 8 columns, as follows:

    global_id: Unique identifier for each entry (int64).
    chunk_name: General category or type of item (object).
    title: Title or name of the item (object).
    description: Description of the item, which may contain HTML tags (object).
    colour: Colour of the item (object).
    eu_size: European size of the item, if applicable (object).
    dropdown_size: Another size attribute, likely in a different format or for different types of items (object, with mixed types warning).
    number_of_products_in_package: Indicates how many items are included in the package, if this is a multi-item package (float64).
